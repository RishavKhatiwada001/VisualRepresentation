6.2 Data Transformation:

1. ”School”, ”sex”, ”address”, ”schoolsup”, ”famsup”, ”activites”, ”nursery”, ”internet”, and ”romantic”
into binary: 0 or 1 (create new columns without overwriting the existing ones).

```python
# creating new column for binary value.
df["school_binary"] = df["school"].replace({"GP": 0, "MS": 1})
df["sex_binary"] = df["sex"].replace({"F": 0, "M": 1})
df["address_binary"] = df["address"].replace({"U": 0, "R": 1})
df["schoolsup_binary"] = df["schoolsup"].replace({"no": 0, "yes": 1})
df["famsup_binary"] = df["famsup"].replace({"no": 0, "yes": 1})
df["activities_binary"] = df["activities"].replace({"no": 0, "yes": 1})
df["nursery_binary"] = df["nursery"].replace({"no": 0, "yes": 1})
df["internet_binary"] = df["internet"].replace({"no": 0, "yes": 1})
df["romantic_binary"] = df["romantic"].replace({"no": 0, "yes": 1})
df
```

Explination:
In the above code, creating new column and where value kept as 0 and 1 by searching for specific value if the value is true putting 0 else putting 1. for example, for school, if the value has GP in it we put 0 in created column school_binary else putting 0.

2. ”Medu”, ”Fedu”, ”reason”, ”guardian”, ”studytime”, ”freetime”, and ”health” into ordinal numbers based
on the number cases in the data set (create news columns without overwriting the existing ones).

```python
#”Medu”, ”Fedu”, ”reason”, ”guardian”, ”studytime”, ”freetime”, and ”health”
# map assign unique values to the column
df["Medu_ordinal"] = df["Medu"].map({"none" : 1, 'higher education':2 ,'primary education (4th grade)':3 ,'secondary education':4,
 '5th to 9th grade':5})
df["Fedu_ordinal"] = df["Fedu"].map({'higher education':1, 'primary education (4th grade)':2, '5th to 9th grade':3,
 'secondary education':4, 'none':5})
df["reason_ordinal"] = df["reason"].map({'course':1, 'other':2, 'home':3, 'reputation':4})
df["guardian_ordinal"] = df["guardian"].map({'mother':1, 'father':2, 'other':3})
df["studytime_ordinal"] = df["studytime"].map({'2 to 5 hours':1, '5 to 10 hours':2, '<2 hours' '>10 hours':3})
df["freetime_ordinal"] = df["freetime"].map({'medium':1, 'low':2, 'high':3, 'very low':4, 'very high':5})
df["health_ordinal"] = df["health"].map({'good':1, 'excellent':2, 'very bad':3, 'bad':4, 'very good':5})
df
```

3. Convert column ”age” to interval datatype. i.e. Create a new column name category age whose values
should be based on the frequency in the column ”age”, You can create categorical data with following
interval.
(a) interval1: [15-17]
(b) interval2: [18-20]
(c) interval3: [21-all]

```python
df["category_age"] = pd.cut(df["age"], bins=[14, 17, 20, df["age"].max()+1], labels=["interval1", "interval2", "interval3"])
df["category_age"]
```

4. Create a new column name passed (yes or no) whose values should be based on the values present in the
G3 column (>= 8–yes, < −no).

```python
df['passed'] = df['G3'].apply(lambda x: 'yes' if x >= 8 else 'no')
df
```

6.3 Descriptive Analysis:

Write a code to show the summary statistic (sum, mean, median, standard deviation, max, and min) of the
variables age, absences, G1, G2, and G3).

```python
# Select the columns to include in the summary statistics
columns = ["age", "absences", "G1", "G2", "G3"]

# Calculating sum, mean, median,maximum value, minimum value, standard deviation.
sum_stats = df[columns].sum()
mean_stats = df[columns].mean()
median_stats = df[columns].median()
std_stats = df[columns].std()
max_stats = df[columns].max()
min_stats = df[columns].min()

# Display the summary statistics
print("Sum:", sum_stats)
print("Mean:", mean_stats)
print("Median:", median_stats)
print("Standard deviation:", std_stats)
print("Max:", max_stats)
print("Min:", min_stats)
```

1. Write a code to show histogram plots and boxplots to visualize the distribution of the variables
”age”, ”absence”, and ”G3”. Interpret the results and comment on the distribution of each variable.
Write your answer in text cell of jupyter notebook.

```python
# create a histogram plot for the variable "age"
plt.hist(df['age'],bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()

# create a histogram plot for the variable "absence"
plt.hist(df['absences'],bins=10)
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.title('Distribution of Absence')
plt.show()

# create a histogram plot for the variable "G3"
plt.hist(df['G3'],bins=10)
plt.xlabel('G3')
plt.ylabel('Frequency')
plt.title('Distribution of G3')
plt.show()

# create a boxplot for the variable "age"
plt.boxplot(df['age'])
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.show()

# create a boxplot for the variable "absence"
plt.boxplot(df['absences'])
plt.ylabel('Absences')
plt.title('Boxplot of Absence')
plt.show()

# create a boxplot for the variable "G3"
plt.boxplot(df['G3'])
plt.ylabel('G3')
plt.title('Boxplot of G3')
plt.show()
```

2. Write a code to show a bar graph of the total number of students who passed the final term grouped
according to the school that they belong to. Use proper labels in the graph and interpret the results.

```python
df.groupby("school")['passed'].value_counts().plot(kind="bar")
```

In a groupby operation, the object is split, a function is applied, and the products are combined. Large amounts of data can be grouped using this, and operations can be computed on these groups. In the given table we can see that the school and its passing rate. from the graph we can see GP has highest amount of student who passed and non gp student than ms student and finally non ms student.

3. Write a code to show a pie chart of analysis of students’ grades as per their school. Use proper labels
in the graph and interpret the results.

```python
df.groupby(['G1','school']).size().unstack().plot.pie(subplots=True,figsize=(10, 10))
df.groupby(['G2','school']).size().unstack().plot.pie(subplots=True,figsize=(10, 10))
df.groupby(['G3','school']).size().unstack().plot.pie(subplots=True,figsize=(10, 10))
```

Above pie chart represent the data of two schools GP and MS. the graph is of g1, g2 and g3(ie. grade1, grade2, and grade3).
